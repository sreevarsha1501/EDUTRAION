```python
# 1. Import Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#2. Load the dataset
d = pd.read_csv('/content/IMDb Movies India.csv', encoding='latin1')
d.head()
#since if we dont use encoding='latin1', we get UnicodeDecodeError
```

```python
#3. Explore the dataset
d.info()
print("\n-------------------------------\n")
print(d.describe())
print("\n-------------------------------\n")
print('Checking the null values: ')
print(d.isnull().sum())
```

```python
print(d.columns)
```

```python
# 4. Select Required Columns
d = d[['Name', 'Year', 'Duration', 'Genre', 'Rating', 'Votes', 'Director', 'Actor 1']]

# Drop rows with missing Rating (our target variable)
d.dropna(subset=['Rating'], inplace=True)

# Fill missing values in other columns with mode
for col in ['Genre', 'Votes', 'Director', 'Actor 1', 'Year', 'Duration']:
    d[col].fillna(d[col].mode()[0], inplace=True)
```

```python
# 5a. Clean and convert 'Votes'
d['Votes'] = d['Votes'].astype(str).str.replace(',', '', regex=True)
d = d[d['Votes'].str.isnumeric()]  # keep numeric only
d['Votes'] = d['Votes'].astype(int)

# 5b. Clean and convert 'Duration'
d['Duration'] = d['Duration'].astype(str).str.replace(' min', '', regex=True).astype(int)

# 5c. Clean and convert 'Year'
d['Year'] = d['Year'].astype(str).str.replace(r'[^\d]', '', regex=True).astype(int)
```
